```python
from bs4 import BeautifulSoup
import requests
```

```python
book = requests.get('https://bookpage.com/reviews')
book
```

```python
book.content
```

```python
soup = BeautifulSoup(book.content)
soup
```

```python
book1 = soup.find('h4',class_='italic')
book1
```

```python
book1.text
```

```python
book1_n = book1.text.replace('\n','')
book1_n
```

```python
name = soup.find_all('h4',class_='italic')
name
```

```python
names = []
for i in name:
    names.append(i.text.replace('\n',''))
names[0:5]
```

```python
genre = soup.find_all('p',class_='genre-links hidden-phone')
genre
```

```python
genre1 = []
for i in genre:
    genre1.append(i.text.replace('\n',''))
genre1[0:5]
```

```python
a_name = soup.find_all('p',class_='sans bold')
a_name
```

```python
author = []
for i in a_name:
    author.append(i.text.replace('\n',''))
author[0:5]
```

```python
page = requests.get('https://bookpage.com/reviews/')
soup = BeautifulSoup(page.content, 'html.parser')
url_tags = soup.find_all("h4",class_="italic")   
url_tags
```

```python
urls = []
for i in url_tags:
     for j in i.find_all("a", href=True):
        if j.text:
            urls.append(j['href'])
urls
```

```python
books_review = []
for url in urls:
        book = requests.get('https://www.bookpage.com'+url)
        soup = BeautifulSoup(book.content, 'html.parser')
        books_review.append(soup.find('div', class_= "article-body").text.replace('\n',''))
books_review
```

```python
import pandas as pd
book_reviews = pd.DataFrame({})
book_reviews['Title(Book_name)'] = names[0:5]
book_reviews['Title(Genre)'] = genre1[0:5]
book_reviews['Title(Author)'] = author[0:5]
book_reviews['Title(Review)'] = books_review[0:5]
book_reviews
```
