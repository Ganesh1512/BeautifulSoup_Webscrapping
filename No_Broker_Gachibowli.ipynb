```python
soup = BeautifulSoup(nb.content)
soup
```

```python
house = soup.find_all('h2',class_='heading-6 font-semi-bold nb__1AShY')
house
```

```python
name = []
for i in house:
    name.append(i.text)
name
```

```python
loc = soup.find_all('div',class_='nb__2CMjv')
loc
```

```python
location=[]
for i in loc:
    location.append(i.text)
location
```

```python
emi = soup.find_all('div',class_='font-semi-bold heading-6')
emi
```

```python
emi1 = []
for i in emi:
    emi1.append(i.text)
emi1
```

```python
emi_pm=[]
for i in range(1,len(emi1),3):
    emi_pm.append(emi1[i].strip())
emi_pm
```

```python
area=[]
for i in range(0,len(emi1),3):
    area.append(emi1[i].strip())
area
```

```python
price=[]
for i in range(2,len(emi1),3):
    price.append(emi1[i].strip())
price
```

```python
import pandas as pd
NB_Gachibowli=pd.DataFrame({})
NB_Gachibowli['House_Title'] = name
NB_Gachibowli['Location'] = location
NB_Gachibowli['Area'] = area
NB_Gachibowli['EMI'] = emi_pm
NB_Gachibowli['Price'] = price
NB_Gachibowli
```
